```python
Medi
```

```python
Medi.drop('자치구',axis=1,inplace=True)
Medi.drop([2,3,21,22,23],inplace=True)
```

```python
Medi.corr()
```

```python
def standardize(df):
    result = df.copy()
    for feature_name in df.columns:
        max = df[feature_name].max()
        min = df[feature_name].min()
        result[feature_name] = 29*((df[feature_name] - min) / (max-min))+1
    return result
Medi_stand = standardize(Medi)
Medi_stand
```

```python
Medi_stand.describe()
```

```python
import statsmodels.api as sm
```

```python
data = Medi_stand.drop(['TotalFertilityRate'],axis=1)
target = Medi_stand[['TotalFertilityRate']]

model = sm.OLS(target,data)
Medi_result = model.fit()
print(Medi_result.summary())
```

```python
Medi_result.params = (Medi_result.params/Medi_result.params.sum())
Medi_stand = Medi_stand.drop('TotalFertilityRate',axis=1)
MediCalcFormula = Medi_stand*Medi_result.params
MediCalcFormula['MediCalcFormula'] = MediCalcFormula.sum(axis=1)
```
